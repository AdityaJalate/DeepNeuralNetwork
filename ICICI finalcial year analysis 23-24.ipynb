```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates
import datetime as dt

!pip install tensorflow keras
```

```python
from tensorflow.keras.layers import Dense, Dropout
```

```python
from tensorflow.keras.layers import GRU
from tensorflow.keras.models import Sequential, load_model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
main_dataset = pd.read_csv("ADANIPORTS.csv") #Read the data
```

```python
main_dataset
```

```python
main_dataset['Date'] = pd.to_datetime(main_dataset['Date'])

# Plot closing prices
plt.plot(main_dataset['Date'], main_dataset['Close'], label="Closing Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Stock Closing Prices Over Time")
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.show()
```

```python
# Create a copy to avoid modifying the original dataset
dataset = main_dataset.copy()

# Convert Date column to datetime format
dataset['Date'] = pd.to_datetime(dataset['Date'])

# Set 'Date' as the index
dataset = dataset.set_index('Date')

# Keep only the 'Close' column
dataset = dataset['Close']  # Use 'Close' instead of 'Cases'

print(dataset.head())  # Check the first few rows
```

```python
# Convert Series back to DataFrame
dataset = dataset.to_frame()

# Rename the column for clarity (optional)
dataset.rename(columns={'Close': 'Price'}, inplace=True)

# Create a moving average column (Expanding Mean with min 3 periods)
dataset['Price_MA2'] = dataset['Price'].expanding(min_periods=3).mean()

# Display first few rows
print(dataset.head())
```

```python
dataset['EMA'] = dataset.iloc[:,0].ewm(span=3,adjust=False).mean() #ewm is used exponential moving average
```

```python
dataset = dataset[dataset['Price_MA2'].notna()]

# Display first few rows to verify
print(dataset.head())
```

```python
dataset.head()
```

```python
# Set 'Close' (renamed as 'Price') as the target variable
target = dataset[['Price']]

dataset = dataset.drop(columns=['Price'])

# Display first few rows
print(dataset.head())
```

```python
# Drop the last row in dataset
dataset = dataset.iloc[:-1]

# Drop the last row in target
target = target.iloc[:-1]

# Display first few rows to verify
print(dataset.tail())
print(target.tail())
```

```python
x, y = dataset.values, target.values
```

```python
x_scaled_set = MinMaxScaler() #Use MinMaxScaler function from sklearn.preprocessing library
y_scaled_set = MinMaxScaler()

X = x_scaled_set.fit_transform(x) #Fit to data and transform the data array x
Y = y_scaled_set.fit_transform(y.reshape(-1,1)) #Fit to data and transform the data array y. Reshape the y variable to long format
```

```python
X_train_data=X[0:77,]
y_train_data=Y[0:77,]
X_test_data=X[78:86,]
y_test_data=Y[78:86,]
```

```python
X_train_data = X_train_data.reshape((-1,1,2)) #Reshape train dataset to required format
X_test_data = X_test_data.reshape((-1,1,2)) #Reshape test dataset to required format
```

```python
from keras.models import Sequential
from keras.layers import GRU, Dense, Dropout

# Initialize the model
GRU_model = Sequential()

# First GRU layer with 512 units
GRU_model.add(GRU(units=512, return_sequences=True, input_shape=(1, dataset.shape[1])))
GRU_model.add(Dropout(0.2))  # Dropout to prevent overfitting

# Second GRU layer with 256 units
GRU_model.add(GRU(units=256))
GRU_model.add(Dropout(0.2))

# Output layer (Regression task - use linear activation)
GRU_model.add(Dense(1))

# Compile the model
GRU_model.compile(loss='mse', optimizer='adam')

# Summary of the model
GRU_model.summary()
```

```python
GRU_model.fit(X_train_data,y_train_data,batch_size=250, epochs=500, validation_split=0.15, verbose=1)
```

```python
# Ensure X_test_data has the correct shape for GRU
print("X_test_data shape:", X_test_data.shape)  # Should be (num_samples, 1, num_features)

# Evaluate the model on test data
score = GRU_model.evaluate(X_test_data, y_test_data)
print('Loss (MSE):', score)

# Predict using the trained model
y_pred = GRU_model.predict(X_test_data)

# Reshape predictions and actual values to match inverse_transform requirements
y_pred = y_pred.reshape(-1, 1)
y_test_data = y_test_data.reshape(-1, 1)

# Apply inverse transformation to get back to original scale
y_test_pred = y_scaled_set.inverse_transform(y_pred)
y_test_actual = y_scaled_set.inverse_transform(y_test_data)

# Plot actual vs predicted values
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(y_test_actual[-100:], label='Actual', color='blue')
plt.plot(y_test_pred[-100:], label='Predicted', color='red')
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.title('Actual vs Predicted Stock Prices')
plt.legend()
plt.show()
```
